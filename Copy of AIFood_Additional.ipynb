```python
from langchain_community.embeddings.fastembed import FastEmbedEmbeddings
```

```python
#embeddings = TextEmbedding(model_name="BAAI/bge-base-en-v1.5")

embeddings = FastEmbedEmbeddings()
```

```python
from langchain_community.vectorstores import FAISS
from langchain_text_splitters import RecursiveCharacterTextSplitter


text_splitter = RecursiveCharacterTextSplitter()
documents = text_splitter.split_documents(docs)
vector = FAISS.from_documents(documents, embeddings)
```

```python
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate
from langchain.llms import Anthropic
import os
from google.colab import userdata

# Get the API key from Colab secrets
anthropic_api_key = userdata.get('claude')

# Initialize the Claude model
llm = Anthropic(anthropic_api_key=anthropic_api_key)

prompt = ChatPromptTemplate.from_template("""Answer the following question based only on the provided context:

<context>
{context}
</context>

Question: {input}""")

document_chain = create_stuff_documents_chain(llm, prompt)
```

```python
from langchain.chains import create_retrieval_chain

retriever = vector.as_retriever()
retrieval_chain = create_retrieval_chain(retriever, document_chain)
```

```python
response = retrieval_chain.invoke({"input": "how can langsmith help with testing?\n\nAssistant:"})
print(response["answer"])


# LangSmith offers several features that can help with testing:...
```

Chat history save

```python
#!pip install --upgrade langchain

from langchain.schema import HumanMessage, SystemMessage, AIMessage
from langchain.prompts.chat import (
    ChatPromptTemplate,
    SystemMessagePromptTemplate,
    HumanMessagePromptTemplate,
    AIMessagePromptTemplate,
    MessagesPlaceholder
)

prompt = ChatPromptTemplate.from_messages([
    ("system", "Answer the user's questions based on the below context:\n\n{context}"),
    MessagesPlaceholder(variable_name="chat_history"),
    ("user", "{input}"),
])
document_chain = create_stuff_documents_chain(llm, prompt)

retrieval_chain = create_retrieval_chain(retriever, document_chain)
```

```python
from langchain_core.messages import HumanMessage, AIMessage
chat_history = [HumanMessage(content="Can LangSmith help test my LLM applications?"), AIMessage(content="Yes!")]
retrieval_chain.invoke({
    "chat_history": chat_history,
    "input":  "her name is 3,  tell  me how langsmith is helpful"
})
```
